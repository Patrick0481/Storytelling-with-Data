```python
import matplotlib.pyplot as plt
import pandas as pd

url = 'https://raw.githubusercontent.com/Patrick0481/Storytelling-with-Data/refs/heads/main/Titanic.csv?token=GHSAT0AAAAAADKNWND7GSSMHKF32VCS53ZS2GBPQ5A' 
titanic = pd.read_csv(url)

# Define children as under 18
children = titanic[titanic['Age'] < 18]
adults = titanic[titanic['Age'] >= 18]

#grouped_data = titanic['Sex'].value_counts()
#grouped_data = titanic.groupby('Sex')['Survived'].sum()
adult_survivors = adults.groupby('Sex')['Survived'].sum()
children_survivors = children['Survived'].sum()


# Create the bar plot
# grouped_data.index creates each category, the grouped_data.values provides the counts
# in other words: the index is the x-axis, the values the y-axis
labels = ['Adult Female', 'Adult Male', 'Children']
values = [adult_survivors.get('female', 0), adult_survivors.get('male', 0), children_survivors]
colors = ['pink', 'blue', 'orange']


plt.bar(labels, values, color=colors)
plt.title('Number of Survivors: Adults by Sex, Children Combined')
plt.ylabel('Number of Survivors')
plt.show()


print("Adults", adult_survivors)

print("Childeren", children_survivors)

print("Total survivors", adult_survivors + children_survivors)

# Display the plot
plt.show()
```
